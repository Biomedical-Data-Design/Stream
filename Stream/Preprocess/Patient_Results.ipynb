```python
import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

os.chdir('/content/drive/My Drive/EICU/eicu-collaborative-research-database-2.0')

cols_to_read_a = ['patientunitstayid', 'actualicumortality']
mortality_df = pd.read_csv("apachePatientResult.csv", usecols=cols_to_read_a)
mortality_df = mortality_df.drop_duplicates(subset='patientunitstayid', keep='first')
mortality_df['actualicumortality'] = mortality_df['actualicumortality'].apply(lambda x: 0 if x == 'EXPIRED' else 1)
mortality_df
```

```python
cols_to_read_p = ['patientunitstayid', 'unitdischargestatus']
patient_df = pd.read_csv("patient.csv", usecols=cols_to_read_p)
patient_df = patient_df.drop_duplicates(subset='patientunitstayid', keep='first')
patient_df['unitdischargestatus'] = patient_df['unitdischargestatus'].apply(lambda x: 0 if x == 'Expired' else 1)
patient_df = patient_df.rename(columns={'unitdischargestatus': 'actualicumortality'})
patient_df
```

```python
all_mortality = pd.concat([mortality_df, patient_df])
all_mortality.sort_values(by=['patientunitstayid'], inplace=True)
all_mortality = all_mortality.drop_duplicates(subset='patientunitstayid', keep='first')
all_mortality
```

```python
folder_path = '/content/drive/My Drive/Colab Notebooks'
file_name = 'mortality_data.csv'
all_mortality.to_csv(os.path.join(folder_path, file_name), index=False)
print(f'Saved to {os.path.join(folder_path, file_name)}')
```
